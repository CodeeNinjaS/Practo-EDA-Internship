```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv(r"C:\Users\sarah\Downloads\heart_disease_health_indicators_BRFSS2015 2 (1).csv")
```

```python

# Defining key variables to compare
key_variables = [
    'HighBP', 'HighChol', 'CholCheck', 'BMI', 'Smoker', 'Stroke', 'Diabetes',
    'PhysActivity', 'Fruits', 'Veggies', 'HvyAlcoholConsump', 'GenHlth',
    'MentHlth', 'PhysHlth', 'DiffWalk', 'Age', 'Education', 'Income'
]

# Creating a function to plot the distribution of variables
def plot_distributions(df, variables, target):
    for var in variables:
        plt.figure(figsize=(12, 6))
        
        # Histogram
        plt.subplot(1, 2, 1)
        sns.histplot(data=df, x=var, hue=target, multiple="dodge", shrink=0.8, kde=True)
        plt.title(f'Histogram of {var} by {target}')
        
        plt.tight_layout()
        plt.show()

# Plot the distributions of key variables
plot_distributions(df, key_variables, 'HeartDiseaseorAttack')
```

```python
# 3. Compare Heart Disease Across Demographic Groups
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv(r"C:\Users\sarah\Downloads\heart_disease_health_indicators_BRFSS2015 2 (1).csv")
```

```python
# Define demographic variables
demographic_vars = ['Age', 'Sex', 'Education', 'Income']

# Creating a function to plot bar charts of heart disease prevalence
def plot_prevalence_bar_charts(df, variables, target):
    for var in variables:
        plt.figure(figsize=(10, 6))
        # Calculate prevalence
        prevalence = df.groupby(var)[target].mean().reset_index()
        # Create bar plot
        sns.barplot(x=var, y=target, data=prevalence)
        plt.title(f'Prevalence of {target} by {var}')
        plt.xlabel(var)
        plt.ylabel(f'Prevalence of {target}')
        plt.xticks(rotation=45)
        plt.show()

# Plot bar charts of heart disease prevalence for demographic variables
plot_prevalence_bar_charts(df, demographic_vars, 'HeartDiseaseorAttack')
```
